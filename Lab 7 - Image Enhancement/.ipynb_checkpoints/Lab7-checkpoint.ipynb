## Task 2

```python
blade = cv2.imread('blade.tif', 0)
plt.imshow(blade, 'gray')
```

```python
horiz_d = cv2.filter2D(blade,-1,np.asarray([[-1,-1,-1],[2,2,2],[-1,-1,-1]]))
plt.imshow(horiz_d, 'gray')
cv2.imwrite('files/task2/horiz_d.png', horiz_d)
```

```python
pos_45_d = cv2.filter2D(blade,-1,np.asarray([[2,-1,-1],[-1,2,-1],[-1,-1,2]]))
plt.imshow(pos_45_d, 'gray')
cv2.imwrite('files/task2/pos_45_d.png', pos_45_d)
```

```python
neg_45_d = cv2.filter2D(blade,-1,np.asarray([[-1,-1,2],[-1,2,-1],[2,-1,-1]]))
plt.imshow(neg_45_d, 'gray')
cv2.imwrite('files/task2/neg_45_d.png', neg_45_d)
```

```python
vert_d = cv2.filter2D(blade,-1,np.asarray([[-1,2,-1],[-1,2,-1],[-1,2,-1]]))
plt.imshow(vert_d, 'gray')
cv2.imwrite('files/task2/vert_d.png', vert_d)
```

```python
dst = cv2.filter2D(blade,-1,np.asarray([[1,1,1],[1,-8,1],[1,1,1]]))
plt.imshow(dst, 'gray')
cv2.imwrite('files/task2/dst.png', dst)
```

### Best Angle
**-45 degree is best suited angle.**

```python
blade_mask = horiz_d.copy()
ret, thresh = cv2.threshold(blade_mask, 250, 255, cv2.THRESH_BINARY)

final = cv2.bitwise_and(blade, thresh, mask=None)
plt.imshow(final, 'gray')
cv2.imwrite('fin.png', final)
```

## Task 3

```python
def apply_filter(img1, img2, kernel, name):
    dst1 = cv2.filter2D(img1,-1,np.asarray(kernel))
    dst2 = cv2.filter2D(img2,-1,np.asarray(kernel))
    fig, ax = plt.subplots(nrows=1, ncols=2)
    ax[0].imshow(dst1, 'gray')
    ax[1].imshow(dst2, 'gray')
    cv2.imwrite(f'files/task3/{name}_1.png', dst1)
    cv2.imwrite(f'files/task3/{name}_2.png', dst2)
```

```python
building = cv2.imread('building.tif', 0)
wirebond = cv2.imread('wirebond.tif', 0)
fig, ax = plt.subplots(nrows=1, ncols=2)
ax[0].imshow(building, 'gray')
ax[1].imshow(wirebond, 'gray')
```

```python
apply_filter(building, wirebond, [[-3,-3,5],[-3,0,5],[-3,-3,5]], 'N')
```
